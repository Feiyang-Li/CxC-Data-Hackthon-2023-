```python
import pandas as pd
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

# import processed datasets 
train = pd.read_csv("../DataSet/labeled_processed_train.csv") # trainig data
valid = pd.read_csv("../DataSet/labeled_processed_valid.csv") # validation data

X_train, y_train = train.iloc[:,2:-1],train.iloc[:,-1] # split the y from other columns
X_valid, y_valid = valid.iloc[:,2:-1],valid.iloc[:,-1]
```

```python
X_train
```

```python

baging = ExtraTreesClassifier(n_estimators=12, max_depth=None, min_samples_split=9, random_state=42,max_leaf_nodes=28)
baging.fit(X_train, y_train)
print(baging.score(X_valid, y_valid))
    #16
```

```python
y_result = baging.predict(X_valid)
print(f1_score(y_valid, y_result))
print(roc_auc_score(y_valid, y_result))
```

```python
result_train = pd.read_csv("../DataSet/unlabeled_processed.csv") # trainig data
result_train
```

```python
X_actual = result_train.iloc[:, 1:]
y_actual = baging.predict(X_actual)
print(y_actual)
```
